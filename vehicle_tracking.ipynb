```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO('yolov8n.pt')
```

```python
video_path = "sample_video.mp4"
```

```python
capture = cv2.VideoCapture(video_path)
```

```python
width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(capture.get(cv2.CAP_PROP_FPS))

output = cv2.VideoWriter("output_vehicle_tracking.mp4", cv2.VideoWriter_fourcc(*"XVID"), fps, (width, height))
```

```python
while capture.isOpened():
    ret, frame = capture.read()
    if not ret:
        break

    results = model.track(frame, persist = True, conf = 0.35, iou = 0.5, tracker = "bytetrack.yaml")
    annotated_frame = results[0].plot()

    output.write(annotated_frame)
    cv2.imshow("Vehicle Tracking with ByteTrack", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

capture.release()
output.release()
cv2.destroyAllWindows()
```
